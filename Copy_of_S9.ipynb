# **1. Póker (V34)**


```python
# Gefið

from random import sample

def pókerhönd():
  stokkur = [(i,j) for i in range(1,5) for j in range(1,14)]
  return sample(stokkur, 5)

def með_tvær_tvennur(hönd):
  # Skilar sönnu ef hönd geymir tvær tvennur
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)                 # búa til mengi
  if len(M) != 3:
    return False                 # það verða að vera 3 gildi
  for spil in M:
    talning = gildi.count(spil)
    if talning > 2: return False # og ekkert spil má koma 3svar
  return True                    # það eru tvær tvennur!

def hlutfall_með_2x2(n):
  talning = 0
  for i in range(n):
    hönd = pókerhönd()
    if með_tvær_tvennur(hönd):
      talning += 1
  return talning/n

hf = hlutfall_með_2x2(100000)
print(f'Hlutfall með tvær tvennur: {hf*100:.1f}%')
```

```python
def hlutfall_með_x(n, x):
  talning = 0
  for i in range(n):
    hönd = pókerhönd()
    if x(hönd):
      talning += 1
  return talning/n
```

```python
#a

def með_tvennu(hönd):
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)

  for spil in M:
    talning = gildi.count(spil)
    if talning == 2:
      return True
  return False

ha = pókerhönd()
print(ha)
print()
mt = með_tvennu(ha)
print(mt)
print()
hlfmt = hlutfall_með_x(100000, með_tvennu)
print(f'Hlutfall með þrennu: {hlfmt*100:.1f}%. Samkvæmt upplýsingum af netinu er rétt gildi um 42.2569%	.')
```

```python
#b

def með_þrennu(hönd):
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)

  for spil in M:
    talning = gildi.count(spil)
    if talning == 3:
      return True
  return False

hb = pókerhönd()
print(hb)
print()
mþ = með_þrennu(hb)
print(mþ)
print()
hlfmþ = hlutfall_með_x(100000, með_þrennu)
print(f'Hlutfall með þrennu: {hlfmþ*100:.1f}%. Samkvæmt upplýsingum af netinu er rétt gildi um 2.1128%.')
```

```python
#c

def með_fullt_hus(hönd):
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)

  if len(M) != 2:
    return False

  t3 = False
  t2 = False
  for spil in M:
    if gildi.count(spil) == 3:
      t3 = True
    if gildi.count(spil) == 2:
      t2 = True
  if (t3 == True and t2 == True):
    return True
  else: 
    return False

hc = pókerhönd()
print(hc)
print()
mfh = með_fullt_hus(hc)
print(mfh)
print()
hTestc = [(2, 5), (1, 5), (3, 5), (1, 2), (2, 2)]
print(með_fullt_hus(hTestc))
print()
hlfmfh = hlutfall_með_x(100000, með_fullt_hus)
print(f'Hlutfall með fullt hús: {hlfmfh*100:.1f}%. Samkvæmt upplýsingum af netinu er rétt gildi um 0.1441%.')
```

```python
#d

def með_lit(hönd):
  sort = [spil[0] for spil in hönd]
  M = set(sort)

  cBlack = True
  cRed = True

  for spil in M:
    if (spil == 1 or spil == 3):
      cBlack = False
    if (spil == 2 or spil == 4):
      cRed = False
  
  if (cBlack == True or cRed == True):
    return True
  else:
    return False


hd = pókerhönd()
print(hd)
print()
ml = með_lit(hd)
print(ml)
print()
hTestd = [(2, 1), (2, 2), (2, 3), (4, 1), (4, 2)]
mlt = með_lit(hTestd)
print(mlt)
print()
hlfml = hlutfall_með_x(100000, með_lit)
print(f'Hlutfall með lit: {hlfml*100:.1f}%. Samkvæmt upplýsingum af netinu er rétt gildi um 5%.')


```
